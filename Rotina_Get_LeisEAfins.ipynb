```python
#Lendo propostas do dia:
def criando_base(path,nomeProposicao,nomeAutores,nomeTemas):
    import pandas as pd
    from datetime import datetime
    proposicao = pd.read_csv(path+nomeProposicao+'.csv', encoding='utf-8', engine="python", sep = ';')
    autores = pd.read_csv(path+nomeAutores+'.csv', encoding='utf-8', engine="python", sep = ';')
    temas = pd.read_csv(path+nomeTemas+'.csv', encoding='utf-8', engine="python", sep = ';')
    meus_temas = temas.query('tema=="Administração Pública" | tema=="Economia" | tema=="Cidades e Desenvolvimento Urbano" | tema=="Direito Civil e Processual Civil" | tema=="Direitos Humanos e Minorias" | tema=="Meio Ambiente e Desenvolvimento Sustentável" | tema=="Estrutura Fundiária" | tema=="Energia, Recursos Hídricos e Minerais" | tema=="Viação, Transporte e Mobilidade" | tema=="Agricultura, Pecuária, Pesca e Extrativismo" | tema=="Indústria, Comércio e Serviços" | tema=="Finanças Públicas e Orçamento"')
    lista = ['Retirado pelo Autor','Perdeu a Eficácia','Perdeu a Eficácia','Devolvida ao Autor','Arquivada']
    minhas_proposicoes = proposicao[~proposicao.ultimoStatus_descricaoSituacao.isin(lista)]
    tipos_not = ['Indicação','Requerimento de Informação','Indicação à Mesa','Consulta']
    minhas_proposicoes = minhas_proposicoes[~minhas_proposicoes.descricaoTipo.isin(tipos_not)]
    proposicoes_temas = minhas_proposicoes.merge(meus_temas,how='left',left_on = 'uri',right_on='uriProposicao')
    proposicoes_temas.dropna(subset='tema',inplace=True)
    proposicoes_temas_autores =proposicoes_temas.merge(autores,how='left',left_on='id',right_on='idProposicao',indicator=True)
    columns_keep=['id','uri','siglaTipo_x','ano_x','codTipo','descricaoTipo','ementa','ementaDetalhada','keywords','dataApresentacao','ultimoStatus_dataHora','codTema','tema','tipoAutor','nomeAutor','siglaPartidoAutor','ultimoStatus_descricaoTramitacao','urlInteiroTeor']
    proposicoes_temas_autores=proposicoes_temas_autores[columns_keep]
    columns_rename={'siglaTipo_x':'Sigla_Proposta','ano_x':'Ano','ultimoStatus_descricaoTramitacao':'Tramitacao_ultimoStatus','urlInteiroTeor':'Link_Teor'}
    proposicoes_temas_autores.rename(columns =columns_rename,inplace=True)
    proposicoes_temas_autores.drop_duplicates(subset='id',inplace=True)
    
    return proposicoes_temas_autores
```

```python
def gerando_report_news_proposicoes(db_atual,db_referencia,pathDbReferencia,pathSaveReport):
    import pandas as pd
    from datetime import datetime
    db_referencia_2=pd.read_csv(pathDbReferencia+db_referencia+'.csv')
    db_today_unique = db_atual[~db_atual['id'].isin(db_referencia_2.id.unique())]
    tema_reduced = {'Administração Pública':'admPub','Direitos Humanos e Minorias':'DH_min','Energia, Recursos Hídricos e Minerais':'Energia',
               'Economia':'Econ','Finanças Públicas e Orçamento':'Fin_orç','Meio Ambiente e Desenvolvimento Sustentável':'Env',
               'Viação, Transporte e Mobilidade':'Mobilidade','Direito Civil e Processual Civil':'DirCiv','Indústria, Comércio e Serviços':'Industria',
               'Agricultura, Pecuária, Pesca e Extrativismo':'Agropec','Cidades e Desenvolvimento Urbano':'City_dev','Estrutura Fundiária':'Fund'}
    db_today_unique['Tema_Reduzido']=db_today_unique.tema.map(tema_reduced)
    
    if db_today_unique.empty==True:
        print('Sem novas proposições!')
    else:
        with pd.ExcelWriter(pathSaveReport+'projetos_'+datetime.today().strftime('%d-%m-%Y')+'.xlsx') as writer:
            for temas in db_today_unique.Tema_Reduzido.unique():
                db_today_unique.query('Tema_Reduzido==@temas').groupby(['Tema_Reduzido','id']).agg(lambda x: x).to_excel(writer,sheet_name=temas+'_'+datetime.today().strftime('%d-%m-%Y'))
        return db_today_unique
    
    

    
    
```

```python
def gerando_avanço_propostas(db_atual ,db_referencia,pathDbReferencia,pathSaveReport):
    import pandas as pd
    from datetime import datetime
    db_referencia_2 = pd.read_csv(pathDbReferencia+db_referencia+'.csv')
    db_referencia_2.ultimoStatus_dataHora=pd.to_datetime(db_referencia_2.ultimoStatus_dataHora).dt.date
    db_atual.ultimoStatus_dataHora=pd.to_datetime(db_atual.ultimoStatus_dataHora).dt.date
    db_avanço_status = db_atual[~db_atual['Tramitacao_ultimoStatus'].isin(db_referencia_2['Tramitacao_ultimoStatus'])]
    db_avanço_hora = db_atual[~db_atual.ultimoStatus_dataHora.isin(db_referencia_2.ultimoStatus_dataHora)]
    my_db_avanco = pd.concat([db_avanço_status, db_avanço_hora], axis=0)
    my_db_avanco.drop_duplicates(subset='id' ,inplace = True)
    tema_reduced = {'Administração Pública':'admPub','Direitos Humanos e Minorias':'DH_min','Energia, Recursos Hídricos e Minerais':'Energia',
               'Economia':'Econ','Finanças Públicas e Orçamento':'Fin_orç','Meio Ambiente e Desenvolvimento Sustentável':'Env',
               'Viação, Transporte e Mobilidade':'Mobilidade','Direito Civil e Processual Civil':'DirCiv','Indústria, Comércio e Serviços':'Industria',
               'Agricultura, Pecuária, Pesca e Extrativismo':'Agropec','Cidades e Desenvolvimento Urbano':'City_dev','Estrutura Fundiária':'Fund'}
    my_db_avanco['Tema_Reduzido']=my_db_avanco.tema.map(tema_reduced)
    
    if my_db_avanco.empty==True:
        print('Sem novos avanços!')
    else:
        with pd.ExcelWriter(pathSaveReport+'projetos_avanço_'+datetime.today().strftime('%d-%m-%Y')+'.xlsx') as writer:
            for temas in my_db_avanco.Tema_Reduzido.unique():
                my_db_avanco.query('Tema_Reduzido==@temas').groupby(['Tema_Reduzido','id']).agg(lambda x: x).to_excel(writer,sheet_name=temas+'_'+datetime.today().strftime('%d-%m-%Y'))
        return my_db_avanco
    
    
```
